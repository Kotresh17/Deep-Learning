# Activation Functions

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def derivative_sigmoid(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def derivative_relu(x):
    return np.log(1/(1+np.exp(x)))
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
def derivative_tanh(x):
    return 1-tanh(x)**2
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y_sigmoid = sigmoid(x)
y_sigmoid_derivative = derivative_sigmoid(x)
y_relu = relu(x)
y_relu_derivative = derivative_relu(x)
y_tanh = tanh(x)
y_tanh_derivative = derivative_tanh(x)
```

```python
plt.plot(x, y_sigmoid, label='sigmoid', color='g', lw=1, ls='-')
plt.plot(x, y_sigmoid_derivative, label='sigmoid_derivative', color='g', lw=1, ls='--')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.show()
```

```python
plt.plot(x, y_relu, label='ReLU', color='g', lw=1, linestyle='-')
plt.plot(x, y_relu_derivative, label='ReLU_derivative', color='g', lw=1, linestyle='--')
plt.legend()
plt.show
```

```python
plt.plot(x, y_tanh, label='tanh', color='g', lw=1, linestyle='-')
plt.plot(x, y_tanh_derivative, label='tanh_derivative', color='g', lw=1, linestyle='--')
plt.legend()
plt.show()
```
